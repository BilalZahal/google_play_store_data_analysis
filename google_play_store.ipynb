* Boşluklar hataya neden olabilir bu yüzden string metodu ile boşlukları _ ile dolduracağız.

* Eksik verileri silmek mantıklı değildir. Bu yüzden eksik veri yerine medyan değerlerini atayacağız. Önce medyan değerlerini bulalım

```python
rating_median = df["Rating"].median()
rating_median
```

* Bulunan medyan değerini fillna ile eskik verilerin yerine atıyorum

```python
df["Rating"].fillna(rating_median, inplace= True)
```

* Diğer sütünlarda ki eksik veriler az olduğu için dropna methodu ile kaldıralarım

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum().sum()
```

```python
df["Reviews"].describe()
```

* Seçili sütünün tipi obje türünde ama int olmalı bunu yapalım şimdi

```python
df["Reviews"] = df["Reviews"].astype("int64")
```

```python
df["Reviews"].describe()
```

* Değerleri birler basamğına göre yuvarlamak için raound methodunu çağırtalım 

```python
df["Reviews"].describe().round()
```

* Bu sütünda aykırı değerler olduğunu görüyoruz

* unique ile tek değerleri görelim

```python
print(len(df["Size"].unique()))
df["Size"].unique()
```

* Rplace methodu ile m ve k değerlerini kaldıralım

```python
df["Size"].replace("M", "", regex=True, inplace=True)
df["Size"].replace("k", "", regex=True, inplace=True)
```

```python
df["Size"].unique()
```

* Değerler arasında metinsel ifade yer alıytor onun yerine medyan değerini yazalım

```python
size_median = df[df["Size"]!= "Varies with device"]["Size"].astype(float).median()
size_median
```

```python
df["Size"].replace("Varies with device", size_median, inplace= True)
```

```python
df.Size = pd.to_numeric(df.Size)
```

```python
df.Size.head()
```

```python
df.Size.describe().round()
```

* Installs sütünündaki tek değerleri unique ile çağıralım

```python
df["Installs"].unique()
```

```python
df.Installs = df.Installs.apply(lambda x:x.replace("+", ""))
df.Installs = df.Installs.apply(lambda x:x.replace(",", ""))
df.Installs = df.Installs.apply(lambda x:int(x))
```

```python
df["Installs"].unique()
```

```python
df["Price"].unique()
```

```python
df.Price = df.Price.apply(lambda x:x.replace("$", ""))
df.Price = df.Price.apply(lambda x:float(x))
```

```python
df["Price"].unique()
```

```python
len(df["Genres"].unique())
```

```python
df["Genres"].head(10)
```

* Alt kategoriler ; ile ayrılmış önce bu değerleri ayıralım

```python
df["Genres"] = df["Genres"].str.split(";").str[0]
```

```python
len(df["Genres"].unique())
```

```python
df["Genres"].unique()
```

* Grupların sayısı için value count methodu kullanalım

```python
df["Genres"].value_counts()
```

```python
df["Genres"].replace("Music & Audio ", "Music", inplace = True)
```

```python
df["Last_Updated"].head()
```

* Obje türünde lki sütünü date time ile değiştirelim

```python
df["Last_Updated"] = pd.to_datetime(df["Last_Updated"])
```
